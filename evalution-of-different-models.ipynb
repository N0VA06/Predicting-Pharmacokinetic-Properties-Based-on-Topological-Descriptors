```python
import matplotlib.pyplot as plt
import seaborn as sns

models = [
    'XGBRegressor', 'Basic NN', 'AdaBoost',
    'KNN Regressor', 'Lasso Regressor', 'Ridge Regressor',
    'Linear Regression', 'Decision Tree'
]

r_squared = [
    0.9363, 0.9509, 0.9192,
    0.8470, 0.9475, 0.9818,
    0.9832, 0.8912
]

rmse = [
    9.2440, 7.9757, 11.6024,
    19.2363, 2.7539, 2.2016,
    1.6139, 12.7918
]

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.barplot(x=r_squared, y=models, palette='viridis')
plt.title('R-squared Scores')
plt.xlabel('R-squared Score')
plt.ylabel('Models')

plt.subplot(1, 2, 2)
sns.barplot(x=rmse, y=models, palette='magma')
plt.title('RMSE Values')
plt.xlabel('RMSE')
plt.ylabel('Models')

plt.tight_layout()
plt.show()
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(models, r_squared, marker='o', linestyle='-', color='b')
plt.title('R-squared Scores')
plt.xlabel('Models')
plt.ylabel('R-squared Score')
plt.xticks(rotation=45)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(models, rmse, marker='o', linestyle='-', color='r')
plt.title('RMSE Values')
plt.xlabel('Models')
plt.ylabel('RMSE')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```
